```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=42)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OrdinalEncoder
```

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer=ColumnTransformer(transformers=[
    ("tnf1",SimpleImputer(),numeric_null),
    ("tnf2",OneHotEncoder(drop='first',sparse_output=False),One_hot_cols)],remainder="passthrough"
)
```

```python
X_train_transform=transformer.fit_transform(X_train)
X_test_transform=transformer.transform(X_test)
```

```python
X_train_transform.shape
```

```python
Labelencoder=LabelEncoder()
```

```python
y_train_transform=Labelencoder.fit_transform(y_train)
y_test_transform=Labelencoder.transform(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale=StandardScaler()
```

```python
X_train_scaled=scale.fit_transform(X_train_transform)
X_test_scaled=scale.transform(X_test_transform)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RF=RandomForestClassifier()
```

```python
RF.fit(X_train_scaled,y_train_transform)
```

```python
#X_train_scaled = pd.DataFrame(X_train_scaled, columns=transformer.get_feature_names_out())   #to convert back into dataframe with column name
#X_test_scaled = pd.DataFrame(X_test_scaled, columns=transformer.get_feature_names_out())
```

```python
y_pred=RF.predict(X_test_scaled)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test_transform,y_pred)
```
